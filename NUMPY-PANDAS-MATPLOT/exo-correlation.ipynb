```python
import numpy as np
x = np.random.randint(1, 5, 10)
print(x)
y = np.array([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])
print(y)

r = np.corrcoef(x, y)
r
```

```python
r = np.corrcoef(x,y)
print(r)
```

## Corr√©lation entre cos(x) et sin(x)

```python
x = np.linspace(0, -300, 1000)
y_cos = np.cos(x)
y_sin = np.sin(x)

r = np.corrcoef(y_cos,y_sin)
print(r)
```

```python
import matplotlib.pyplot as plt

x = np.random.normal(-100, 100, 50)
y = np.random.normal(-100, 100, 50)
z = np.random.normal(-100, 100, 50)

print(np.corrcoef([x,y,z]))

plt.plot(x)
plt.plot(y)
plt.plot(z)
```
