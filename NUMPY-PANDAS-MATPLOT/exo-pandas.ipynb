```python
# Étape 1: Importation des bibliothèques et chargement des fichiers CSV

import pandas as pd

# Charger le fichier de données "par sexe"
data_sexe = pd.read_csv('/Users/cedriclag/Desktop/exercices_simplon/csv/covid-hospit-2023-03-31-18h01.csv', sep=';', quotechar='"')

# Charger le fichier contenant la population par département
pop_dep = pd.read_csv('/Users/cedriclag/Desktop/exercices_simplon/csv/pop-dep-france.csv', sep=';', header=None, names=['dep', 'departement', 'population'])
# Supprimer les lignes où la population est "Total"
pop_dep = pop_dep[pop_dep['population'] != 'Total']

# Convertir la colonne population en int en supprimant les espaces insécables
pop_dep['population'] = pop_dep['population'].str.replace('\u202f', '').astype(int)

# Vérifiez les premières lignes pour vous assurer que les données ont été correctement traitées
print(pop_dep.head())
```

```python
# Étape 2: Fusion des données et calcul du taux de mortalité

# Fusionner les données sur le numéro de département
merged_data = pd.merge(data_sexe[data_sexe['sexe'] == 0], pop_dep, left_on='dep', right_on='dep', how='left')

# Calculer le taux de mortalité
merged_data['mortality_rate'] = (merged_data['dc'] / merged_data['population']) * 100

# Vérifiez les premières lignes pour vous assurer que la fusion a été correctement effectuée
print(merged_data[['dep', 'mortality_rate']].head())
```

```python
# Étape 2.1: Vérification de la fusion

# Vérifiez les numéros de département présents dans data_sexe et pop_dep
deps_data_sexe = set(data_sexe['dep'].unique())
deps_pop_dep = set(pop_dep['dep'].unique())

# Identifiez les numéros de département qui ne sont présents que dans l'un des deux DataFrames
only_in_data_sexe = deps_data_sexe - deps_pop_dep
only_in_pop_dep = deps_pop_dep - deps_data_sexe

print("Départements présents seulement dans data_sexe:", only_in_data_sexe)
print("Départements présents seulement dans pop_dep:", only_in_pop_dep)
```

```python
# Étape 2.2: Harmonisation des formats de département

# Convertissez les numéros de département dans pop_dep au format de deux chiffres
pop_dep['dep'] = pop_dep['dep'].apply(lambda x: x.zfill(2))

# Supprimez les lignes inutiles de pop_dep
rows_to_remove = ['France métropolitaine ', 'France métropolitaine et DOM', 'DOM']
pop_dep = pop_dep[~pop_dep['dep'].isin(rows_to_remove)]

# Refusionner les données
merged_data = pd.merge(data_sexe[data_sexe['sexe'] == 0], pop_dep, left_on='dep', right_on='dep', how='left')

# Recalculer le taux de mortalité
merged_data['mortality_rate'] = (merged_data['dc'] / merged_data['population']) * 100

# Vérifiez les premières lignes pour vous assurer que la fusion a été correctement effectuée
print(merged_data[['dep', 'mortality_rate']].head())
```

```python
# Étape 3: Tracer l'histogramme du taux de mortalité par département

import matplotlib.pyplot as plt

# Triez le DataFrame merged_data par taux de mortalité décroissant
sorted_data = merged_data.sort_values(by='mortality_rate', ascending=False)

# Utilisez Matplotlib pour tracer l'histogramme
plt.figure(figsize=(20, 10))
plt.bar(sorted_data['dep'], sorted_data['mortality_rate'], color='skyblue')
plt.xlabel('Numéro de département')
plt.ylabel('Taux de mortalité (%)')
plt.title('Taux de mortalité du Covid par département en France')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Étape 4: Trier par numéro de département et tracer l'histogramme

# Triez le DataFrame merged_data par numéro de département
sorted_by_dep = merged_data.sort_values(by='dep')

# Utilisez Matplotlib pour tracer l'histogramme
plt.figure(figsize=(20, 10))
plt.bar(sorted_by_dep['dep'], sorted_by_dep['mortality_rate'], color='skyblue')
plt.xlabel('Numéro de département')
plt.ylabel('Taux de mortalité (%)')
plt.title('Taux de mortalité du Covid par numéro de département en France')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Étape 4 (Corrigée): Trier par numéro de département et tracer l'histogramme

# Convertir les numéros de département qui sont numériques en entiers, sinon les conserver en tant que chaînes
merged_data['dep_int'] = pd.to_numeric(merged_data['dep'], errors='coerce')
merged_data = merged_data.sort_values(by=['dep_int', 'dep'])

# Utilisez Matplotlib pour tracer l'histogramme
plt.figure(figsize=(20, 10))
plt.bar(merged_data['dep'], merged_data['mortality_rate'], color='skyblue')
plt.xlabel('Numéro de département')
plt.ylabel('Taux de mortalité (%)')
plt.title('Taux de mortalité du Covid par numéro de département en France')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
